```python
# create the polytopes
cube = polytopes.hypercube(3)
cube.plot()

# ico = polytopes.icosahedron()
# ico.plot()
```

```python
# extract the interface between two half-spaces
hspace_a = Polyhedron(ieqs=[[1/2, 1, 0, 0]])
hspace_b = Polyhedron(ieqs=[[-1/2, -1, 0, 0]])

interface_2hspace = hspace_a.intersection(hspace_b)
interface_2hspace.plot()
```

```python
# extract the interface between two half-volumn
def my_intersection(hspace_a, hspace_b):
    hcube_a = cube.intersection(hspace_a)
    hcube_b = cube.intersection(hspace_b)
    interface = hcube_a.intersection(hcube_b)
    return hcube_a, hcube_b, interface

hcube_a, hcube_b, interface_2hcube = my_intersection(hspace_a, hspace_b)

interface_2hcube.plot()
# interface_2hcube.vertices()
```

```python
# extract the interface where a small cube intersect with a big cube
cube_small = polytopes.cube(intervals=[[0, 1], [0, 1], [0, 1]])
cube_big = polytopes.cube(intervals=[[-1, 2/2], [0, 1/2], [0, 1001234/1000000]])
interface_2adjcube = cube_small.intersection(cube_big)

interface_2adjcube.plot()
```
